```python
import pandas as pd

def cleanExcel(excel_file, csv_file):
    try:
        df = pd.read_excel(excel_file)

        for column in df.columns:
            df[column] = df[column].apply(
                lambda x: ' '.join(str(x).split()).strip() if pd.notna(x) else ''
            )

        # make the columns into string
        df = df.astype(str)

        df.to_csv(csv_file, index=False)
        print(f"Successfully converted '{excel_file}' to '{csv_file}' with clean formatting.")

    except FileNotFoundError:
        print(f"Error: '{excel_file}' was not found.")
    except Exception as e:
        print(f"An unexpected error occurred: {e}")

if __name__ == '__main__':
    cleanExcel('FA25_registrations.xlsx', 'FA25_registrations.csv')
```
